```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t
from scipy.stats import binom
```

# 1st task

```python
var = 25
n = 64
mean = 102

h_0 = 100
```

```python
se = np.sqrt(var / n)
se
```

```python
z = (mean - h_0) / se
z
```

```python
p_one_tailed = 1 - norm.cdf(z)
p_one_tailed
```

```python
z_values = np.linspace(-4, 4, 1000)
pdf = norm.pdf(z_values)

plt.plot(z_values, pdf, label='Стандартное нормальное распределение')

# Отмечаем Z-статистику
plt.axvline(z, color='blue', linestyle='--', label=f'Z-статистика = {z:.2f}')

plt.legend()
plt.title('Стандартное нормальное распределение и Z-статистика')
plt.xlabel('Z')
plt.ylabel('Плотность вероятности')
plt.show()
```

# 8th task

```python
def independent_t_test(m0, m1, std0, std1, n, two_tailed=True):
    df = 2 * n - 2
    var_p = ((n - 1) * std0 ** 2 + (n - 1) * std1 ** 2) / df
    se = np.sqrt(var_p * 2 / n)
    t_stat = (m0 - m1) / se

    p_one = 1 - t.cdf(abs(t_stat), df)
    p_value = 2 * p_one if two_tailed else p_one

    return t_stat, p_value
```

```python
_, p = independent_t_test(8, 7.5, 0.5, 0.4, 30, two_tailed=False)
p
```

```python
p * 10000
```
